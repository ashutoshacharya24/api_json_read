```python
import requests
import pandas as pd

def fetch_and_flatten(api_url):
    try:
        # Step 1: Fetch data from API
        response = requests.get(api_url)
        response.raise_for_status()  # Raise error for bad response
  
        data = response.json()  # Convert to JSON
        
        # Step 2: Flatten JSON and fetch all columns
        # If API returns a list of records
        if isinstance(data, list):
            df = pd.json_normalize(data, sep='_', max_level=None)
        
        # If API returns nested dictionary
        elif isinstance(data, dict):
            df = pd.json_normalize(data, sep='_', max_level=None)
        
        else:
            raise ValueError("Unsupported JSON structure")
        
        return df
    
    except Exception as e:
        print(f"Error: {e}")
        return None


# Example usage
api_url = "https://jsonplaceholder.typicode.com/users"
df = fetch_and_flatten(api_url)
spark_df = spark.createDataFrame(df)

display(spark_df)
```
